```python
from bs4 import BeautifulSoup as soup 
import pandas as pd
import pymongo 
import datetime
```

```python
# Initialize Mongo 
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.test_raw_html_db
collection = db.job_detail
```

```python
# Load data from our job detail collection in our raw_html_db database
data = pd.DataFrame(list(collection.find().limit(200)))
```

```python
job_link_collection = db.j
pd.DataFrame(list(job_link_collection.find().limit(200)))
```

```python
# create a list of lists that represent each row in our link src column of our data variable
# link_list = []
# for i in data.link_src:
#     for j in i:
# #       convert the raw html into a beautiful soup object we can parse later  
#         link_list.extend(soup(j,'html.parser'))

link_list = []
for html in data.link_src:
    link_list.append(soup(html,'html.parser'))

# def parse_data(data_table):
#     for html in data_table:
#         print(html)

# iterate through each sublist of html, parse it, save the result in a tuple then append it to a list as a dict
headers = []
job_titles = []
senority = []
posted = []
employment_type = []

for idx in range(len(link_list)):  
    for x in link_list[idx].find_all('h3','jobs-top-card__company-info t-14 mt1'):
        vitals = x.get_text().replace('Company Name','').replace('Company Location','').split('\n')
        vitals = tuple([elem for elem in vitals if elem.strip()])
        headers.append({'company':vitals[0].strip(),'location':vitals[1].strip()})
    try:
        for title in link_list[idx].find('h1',"jobs-top-card__job-title t-24"):
            job_titles.append(title)
    except:
        job_titles.append('n/a')
        print(idx,'NOOOOO')
    
    try:
        for x in link_list[idx].find('p','jobs-box__body js-formatted-exp-body'):
            senority.append(x)
    except:
        senority.append('n/a')
        
  

    try:
        for x in link_list[idx].find_all('p','mt1 full-width flex-grow-1 t-14 t-black--light'):
            posted.append(x.find_all('span')[1].text)
    except:
        posted.append('n/a')

    try:
        for x in link_list[idx].find('p','jobs-box__body js-formatted-employment-status-body'):
            employment_type.append(x)
    except:
        employment_type.append('n/a')

df1 = pd.DataFrame(headers)
df2 = pd.DataFrame(job_titles,columns=['job_title'])
df3= pd.DataFrame(senority,columns=['senority'])
df4= pd.DataFrame(posted,columns=['posted_time'])
df5 = pd.DataFrame(employment_type,columns=['employment_type'])
dataset  = pd.concat([df2,df1,df3,df4,df5],axis=1)
dataset['scrape_time'] = datetime.datetime.now()
dataset.head()
```
